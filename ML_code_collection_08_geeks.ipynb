# **Code: Python code for KMean++ Algorithm **

The code you provided implements the K-Means++ algorithm for initializing cluster centers. Here's a step-by-step explanation of the code:

2. Define the `kmeans_pp` function:
```python
def kmeans_pp(X, k):
    centers = []
    # Step 1: Randomly select the first cluster center
    first_center_idx = np.random.choice(len(X))
    centers.append(X[first_center_idx])
    
    while len(centers) < k:
        # Step 2: Compute the minimum distance squared (D^2) from each point to the nearest cluster center
        dist_squared = np.array([min([np.linalg.norm(x - c)**2 for c in centers]) for x in X])
        
        # Step 3: Normalize the distances to obtain probabilities
        probabilities = dist_squared / dist_squared.sum()
        
        # Step 4: Randomly select a new cluster center based on the probabilities
        new_center_idx = np.random.choice(len(X), p=probabilities)
        centers.append(X[new_center_idx])
    
    return np.array(centers)
```

The `kmeans_pp` function takes two arguments: `X` (the dataset) and `k` (the number of clusters). It initializes an empty list `centers` to store the cluster centers.

Inside the function, the K-Means++ algorithm is implemented in the following steps:

- Step 1: Randomly select the first cluster center from the dataset `X`.
- Step 2: For each point in `X`, compute the squared distance to the nearest cluster center already chosen (`dist_squared`).
- Step 3: Normalize the distances `dist_squared` to obtain probabilities that sum up to 1.
- Step 4: Randomly select a new cluster center from `X` based on the probabilities calculated in Step 3.

These steps are repeated until `k` cluster centers are selected. The resulting cluster centers are stored in the `centers` list and returned as a NumPy array.

3. Example usage:

# **Agglomerative Clustering**

Certainly! I'll provide you with the Python code for Agglomerative Clustering and explain each step along with its purpose. Agglomerative Clustering is a hierarchical clustering algorithm that starts with each data point as its own cluster and progressively merges the closest clusters until the desired number of clusters is obtained. Here's the code:

```python
import numpy as np
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import linkage, fcluster
import matplotlib.pyplot as plt


def agglomerative_clustering(X, k):
    # Step 1: Compute the linkage matrix using the complete linkage method
    Z = linkage(X, method='complete')
    
    # Step 2: Perform clustering and obtain cluster labels
    clusters = fcluster(Z, k, criterion='maxclust')
    
    return clusters


# Example usage
# Generate a random dataset with 100 samples and 2 features
X, _ = make_blobs(n_samples=100, n_features=2, centers=3, random_state=42)

# Apply Agglomerative Clustering

cluster_labels = agglomerative_clustering(X, k)

# Plot the clusters

plt.title("Agglomerative Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Now let's break down the steps of the Agglomerative Clustering algorithm and explain their purposes:

1. **Step 1**: Compute the linkage matrix using the complete linkage method. The linkage matrix represents the pairwise distances between the data points and is used to determine which clusters to merge at each step. The "complete" linkage method considers the maximum distance between points in different clusters when calculating the linkage.

2. **Step 2**: Perform clustering and obtain cluster labels. The `fcluster` function is used to assign cluster labels based on the linkage matrix. The `k` parameter specifies the desired number of clusters, and the `criterion='maxclust'` argument ensures that the algorithm stops merging clusters when the desired number of clusters is reached.

Finally, the code plots the resulting clusters using the cluster labels obtained from the Agglomerative Clustering algorithm.

# **code for devisive clustering**

Divisive clustering, also known as Top-Down clustering, is an iterative hierarchical clustering algorithm that starts with all data points in a single cluster and recursively splits them into smaller clusters until the desired number of clusters is achieved. Here's the Python code for Divisive Clustering:

```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt


class DivisiveClustering:
    def __init__(self, k):
        self.k = k
        self.clusters = []

    def fit(self, X):
        self._divide_cluster(X)

    def _divide_cluster(self, X):
        # Step 1: Initialize the first cluster with all data points
        self.clusters.append(X)

        while len(self.clusters) < self.k:
            # Step 2: Select the largest cluster to split
            largest_cluster = max(self.clusters, key=len)

            # Step 3: Split the largest cluster into two smaller clusters
            cluster1, cluster2 = self._split_cluster(largest_cluster)

            # Step 4: Replace the largest cluster with the two smaller clusters
            self.clusters.remove(largest_cluster)
            self.clusters.extend([cluster1, cluster2])

    def _split_cluster(self, cluster):
        # Step 3a: Find the dimension with the largest spread
        max_spread_dim = np.argmax(np.ptp(cluster, axis=0))

        # Step 3b: Sort the cluster based on the selected dimension
        sorted_cluster = cluster[cluster[:, max_spread_dim].argsort()]

        # Step 3c: Split the cluster into two equal-sized subclusters
        midpoint = len(sorted_cluster) // 2
        cluster1 = sorted_cluster[:midpoint]
        cluster2 = sorted_cluster[midpoint:]

        return cluster1, cluster2


# Example usage
# Generate a random dataset with 100 samples and 2 features
X, _ = make_blobs(n_samples=100, n_features=2, centers=3, random_state=42)

# Apply Divisive Clustering
k = 3  # Number of clusters

    plt.scatter(cluster[:, 0], cluster[:, 1], color=colors[i])

```

Let's break down the steps of the Divisive Clustering algorithm and explain their purposes:

1. **Step 1**: Initialize the first cluster with all data points. In this step, we start with a single cluster that contains all the data points.

2. **Step 2**: Select the largest cluster to split. At each iteration, we identify the largest cluster among the existing clusters.

3. **Step 3**: Split the largest cluster into two smaller clusters.
   - **Step 3a**: Find the dimension with the largest spread. We determine the dimension (feature) with the highest range of values in the largest cluster.
   - **Step 3b**: Sort the cluster based on the selected dimension. We sort the data points in the largest cluster based on the selected dimension in ascending order.
   - **Step 3c**: Split the cluster into two equal-sized subclusters. We divide the sorted cluster into two equal-sized subclusters by selecting the midpoint.

4. **Step 4**: Replace the largest cluster with the two smaller clusters. We remove the largest cluster from the list of clusters and replace it with the two smaller subclusters obtained from the split.

The algorithm repeats steps 2-4 until the desired number of clusters (`k`) is reached. Finally, the code plots the resulting clusters obtained from the Divisive Clustering algorithm.

# Reinforcement learning

````python
import gym
import numpy as np

# Define the Q-table and learning rate
q_table = np.zeros((state_size, action_size))
alpha = 0.8
gamma = 0.95

# Train the Q-Learning algorithm
for episode in range(num_episodes):
	state = env.reset()
	done = False
	while not done:
		# Choose an action
		action = np.argmax(
			q_table[state, :] + np.random.randn(1, action_size) * (1. / (episode + 1)))

		# Take the action and observe the new state and reward
		next_state, reward, done, _ = env.step(action)

		# Update the Q-table
		q_table[state, action] = (1 - alpha) * q_table[state, action] + \
			alpha * (reward + gamma * np.max(q_table[next_state, :]))

		state = next_state

# Test the trained Q-Learning algorithm
state = env.reset()
done = False
while not done:
	# Choose an action
	action = np.argmax(q_table[state, :])

	# Take the action
	state, reward, done, _ = env.step(action)
	env.render()
````

what is Q-Learning in Python
----------------------------------



Q-learning is a reinforcement learning algorithm that allows an agent to learn optimal actions in a Markov decision process (MDP) by estimating the values of state-action pairs. It is a model-free approach, meaning it does not require knowledge of the underlying dynamics of the environment.

In Q-learning, the agent maintains a table called the Q-table, which stores the estimated values (Q-values) for each state-action pair. Initially, the Q-values are initialized arbitrarily. The agent interacts with the environment by taking actions and observing the resulting rewards and next states. It updates the Q-values based on the observed rewards and the estimated future rewards.

# **T- test**

````python
from scipy import stats
import numpy as np

# Generate two sets of random data
np.random.seed(0)
group1 = np.random.normal(loc=0, scale=1, size=100)
group2 = np.random.normal(loc=2, scale=1, size=100)

# Perform independent t-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Print the results
print("T-Test Results:")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
````

The t-test is used to compare the means of two groups and determine if they are significantly different. In this example, we generate two sets of random data, group1 and group2. The ttest_ind() function from scipy.stats is used to perform an independent t-test. It returns the t-statistic and the p-value. A low p-value indicates a significant difference between the two groups.

# **chi-square test**

````python
from scipy import stats
import numpy as np

# Create observed and expected frequency arrays
observed = np.array([10, 15, 5])
expected = np.array([10, 10, 10])

# Perform chi-square test
chi_statistic, p_value = stats.chisquare(observed, expected)

# Print the results
print("Chi-Square Test Results:")
print("Chi-Square Statistic:", chi_statistic)
print("P-Value:", p_value)
````

The chi-square test is used to determine if there is a significant association between categorical variables. In this example, we have observed and expected frequency arrays. The chisquare() function from scipy.stats is used to perform a chi-square test. It returns the chi-square statistic and the p-value. A low p-value suggests a significant association between the variables.

# **ANOVA test**

````python
from scipy import stats
import numpy as np

# Generate three sets of random data
np.random.seed(0)
group1 = np.random.normal(loc=0, scale=1, size=100)
group2 = np.random.normal(loc=2, scale=1, size=100)
group3 = np.random.normal(loc=5, scale=1, size=100)

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Print the results
print("ANOVA Test Results:")
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)
````

ANOVA (Analysis of Variance) is used to compare the means of more than two groups. In this example, we generate three sets of random data: group1, group2, and group3. The f_oneway() function from scipy.stats is used to perform a one-way ANOVA. It returns the F-statistic and the p-value. A low p-value indicates a significant difference in means between the groups.
