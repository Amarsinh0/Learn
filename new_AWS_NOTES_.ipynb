# 15) NACL vs. Security Group:

Network Access Control List (NACL):
- Definition: NACL is a stateless firewall at the subnet level, controlling inbound and outbound traffic based on rules you define.
- Example: You can set NACL rules to allow or deny specific IP addresses or ports for incoming or outgoing traffic.

Security Group:
- Definition: Security Groups are stateful firewalls that control inbound and outbound traffic at the instance level.
- Example: You can configure security groups to allow HTTP and SSH traffic to your EC2 instances.

Difference:
The main difference between NACL and Security Group lies in their scope and statefulness. NACL applies at the subnet level and is stateless, meaning you need to explicitly allow inbound and outbound traffic for both directions. Security Groups, on the other hand, apply at the instance level and are stateful, meaning any response to allowed inbound traffic is automatically allowed for the corresponding outbound traffic.

# 24. Question: How can you secure data at rest in Amazon S3?

Answer: You can secure data at rest in Amazon S3 using server-side encryption. S3 provides three options for server-side encryption:

- Server-Side Encryption with Amazon S3 Managed Keys (SSE-S3): S3 automatically encrypts objects using AES-256 encryption and handles key management.
- Server-Side Encryption with AWS Key Management Service (SSE-KMS): S3 uses AWS KMS to manage encryption keys, providing more control and auditability over the encryption process.
- Server-Side Encryption with Customer-Provided Keys (SSE-C): Customers can provide their encryption keys, and S3 uses them to encrypt and decrypt objects.

# 25. Question: What is the difference between Amazon RDS and Amazon Aurora?

Answer: Amazon RDS (Relational Database Service) is a managed database service that supports various database engines like MySQL, PostgreSQL, Oracle, and SQL Server. Amazon Aurora is a fully managed relational database engine compatible with MySQL and PostgreSQL, offering higher performance, scalability, and durability compared to standard RDS databases. Aurora is designed for applications that require high availability and high-performance databases.

- Encrypting data in transit using SSL/TLS for communication between resources in the VPC.
- Encrypting data at rest using services like AWS Key Management Service (KMS) or transparent data encryption for databases.

# 27. Question: How can you ensure high availability for a database in Amazon RDS?

Multi-AZ deployment creates a synchronous replica of the primary database in a different Availability Zone. If the primary database becomes unavailable due to hardware failure, maintenance, or other issues, RDS automatically fails over to the standby replica, ensuring minimal downtime.
