# **MISSING VALUES**

**let us capture all the nan value**

```python
#first lests handel the categorical feature which are meissing
feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype=='O']
for feature in feature_nan:
  print("{}:  {}% missing values". format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

**now replacing new values with new labels**

```python
def replace_cat_feature(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna('missing')
    return data
dataset=replace_cat_feature(dataset,feature_nan)
dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

**now lets check for numerical variables the contains missing values**

```python
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']
# we will print the numerical nan variables and percenatge of missing values
for feature in numerical_with_nan:
  print('{}:{} % missing value'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

### **replacing the numerical missisng value**

```python
for feature in numerical_with_nan:
 
 # we will rplacing by using  median since there are outliers
 median_values=dataset[feature].median()

 #creating a new feature to capture nan values
 dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
 dataset[feature].fillna(median_values,inplace=True)
dataset[numerical_with_nan].isnull().sum()
```

```python
dataset.head()
```

# **TEMPORAL VARIABLES(DATE TIME VARIABLES)**

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
  dataset[feature]=dataset['YrSold']-dataset[feature]

dataset.head()
```

# **Handling rare categoriacl feature**

---

we will remove categorical variables that are present less than 1% of observations

```python
categorical_feature=[feature for feature in dataset .columns if dataset [feature].dtype=='O']
categorical_feature
```

```python
for feature in categorical_feature:
  temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
  temp_df=temp[temp>0.01].index
  dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_Var')
```
